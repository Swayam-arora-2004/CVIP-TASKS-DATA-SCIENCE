```python
daily_data = daily_data.replace('s', pd.NA)
hourly_data = hourly_data.replace('s', pd.NA)
monthly_data = monthly_data.replace('s', pd.NA)
three_hour_data = three_hour_data.replace('s', pd.NA)
```

```python
daily_data = daily_data.replace('T', pd.NA)
hourly_data = hourly_data.replace('T', pd.NA)
monthly_data = monthly_data.replace('T', pd.NA)
three_hour_data = three_hour_data.replace('T', pd.NA)
```

```python
daily_data = daily_data.dropna()
hourly_data = hourly_data.dropna()
monthly_data = monthly_data.dropna()
three_hour_data = three_hour_data.dropna()
```

```python
daily_data = daily_data[daily_data['DailyPeakWindDirection'] != 'VRB']
hourly_data = hourly_data[hourly_data['HourlyWindDirection'] != 'VRB']
three_hour_data = three_hour_data[three_hour_data['HourlyWindDirection'] != 'VRB']
```

```python
daily_data = daily_data.dropna()
daily_data = daily_data[daily_data['REPORT_TYPE'].str.strip() == 'SOD']
hourly_data = hourly_data.dropna()
hourly_data = hourly_data[hourly_data['REPORT_TYPE'].str.strip() == 'FM-15']
monthly_data = monthly_data.dropna()
monthly_data = monthly_data[monthly_data['REPORT_TYPE'].str.strip() == 'SOM']
three_hour_data = three_hour_data.dropna()
three_hour_data = three_hour_data[three_hour_data['REPORT_TYPE'].str.strip() == 'FM-12']
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(daily_data['DailyAverageDryBulbTemperature'], bins=20, kde=True, color='blue')
plt.title('Distribution of Daily Average Temperature')
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MonthlyMeanTemperature', y='MonthlyTotalLiquidPrecipitation', data=monthly_data)
plt.title('Relationship between Monthly Mean Temperature and Total Precipitation')
plt.xlabel('Mean Temperature (°F)')
plt.ylabel('Total Precipitation (inches)')
plt.show()
```

```python
numeric_columns = hourly_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = hourly_data[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Hourly Data')
plt.show()
numeric_columns = monthly_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = monthly_data[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Monthly Data')
plt.show()
numeric_columns = daily_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = daily_data[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Daily Data')
plt.show()
numeric_columns = three_hour_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = three_hour_data[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Three Hourly Data')
plt.show()
```
